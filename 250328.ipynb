```python
score_str = input("시험 점수 (0~100)를 입력하세요: ")
try:
    score = int(score_str)
    if 0 <= score <= 100:
        if score >= 90:
            grade = 'A'
        elif score >= 80:
            grade = 'B'
        elif score >= 70:
            grade = 'C'
        elif score >= 60:
            grade = 'D'
        else:
            grade = 'F'
        print(f"당신의 학점은 {grade}입니다.")
    else:
        print("점수는 0부터 100 사이의 값이어야 합니다.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

```python
def calculate_grade(score):
    """Calculates the grade based on the given score."""
    if 90 <= score <= 100:
        return 'A'
    elif 80 <= score < 90:
        return 'B'
    elif 70 <= score < 80:
        return 'C'
    elif 60 <= score < 70:
        return 'D'
    else:
        return 'F'

try:
    score_str = input("시험 점수 (0~100)를 입력하세요: ")
    score = int(score_str)
    if 0 <= score <= 100:
        grade = calculate_grade(score)
        print(f"당신의 학점은 {grade}입니다.")
    else:
        print("점수는 0부터 100 사이의 값이어야 합니다.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

```python
# 문제 1: 1부터 N까지의 합 계산하기
n_str = input("양의 정수 N을 입력하세요: ")
try:
    n = int(n_str)
    if n <= 0:
        print("N은 양의 정수여야 합니다.")
    else:
        total = 0
        for i in range(1, n + 1):
            total += i
        print(f"1부터 {n}까지의 합은: {total} 입니다.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

```python
n = int(input("양의 정수 N을 입력하세요: "))  # Convert input to an integer
total = 0
for i in range(1, n + 1):
  total += i
  print(f"{i}번: 1번 부터 {n}까지의 힘은: {total}입니다. ")
```
